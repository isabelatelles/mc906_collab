```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_training_loss(history, epochs):
    plt.style.use("ggplot")
    plt.plot(np.arange(0, epochs), history['loss'])
    plt.plot(np.arange(0, epochs), history['val_loss'])
    plt.title('Training Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['train_loss', 'val_loss'], loc='upper left')
    plt.show()
    
def plot_accuracy_loss(history, epochs):
    plt.style.use("ggplot")
    plt.plot(np.arange(0, epochs), history['acc'])
    plt.plot(np.arange(0, epochs), history['val_acc'])
    plt.title('Accuracy Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['train_acc', 'val_acc'], loc='upper left')
    plt.show()
```

# 1st Training
**Baseline Model:** ResNet50
<br>
**Optimizer:** SGD, with learning rate 0.0001
<br>
**Image shape:** 299x299x3
<br>
**Validation Accuracy:** 0.0106
<br>
**Training Accuracy:** 0.5154

```python
with open('dataset2/resnet_sgd.pkl', 'rb') as f:
    history = pickle.load(f)
history
```

```python
EPOCHS = 10
```

```python
plot_training_loss(history, EPOCHS)
```

```python
plot_accuracy_loss(history, EPOCHS)
```
