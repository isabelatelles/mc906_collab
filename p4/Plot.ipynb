```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# First Training
**Baseline Model:** ResNet50
<br>
**Optimizer:** RMSprop, with learning rate 0.0001
<br>
**Image shape:** 224x224x3
<br>
**Validation Accuracy:** 0.7095
<br>
**Training Accuracy:** 0.7730

```python
with open('history.pkl', 'rb') as f:
    history = pickle.load(f)
```

```python
EPOCHS = 10
```

```python
plt.style.use("ggplot")
plt.plot(np.arange(0, EPOCHS), history['loss'])
plt.plot(np.arange(0, EPOCHS), history['val_loss'])
plt.title('Training Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train_loss', 'val_loss'], loc='upper left')
plt.show()
```

```python
plt.style.use("ggplot")
plt.plot(np.arange(0, EPOCHS), history['acc'])
plt.plot(np.arange(0, EPOCHS), history['val_acc'])
plt.title('Training Accuracy')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train_acc', 'val_acc'], loc='upper left')
plt.show()
```
